```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import itertools

# Replace 'Education' and 'Income' with your actual column names
X = data['Education']
y = data['Income']

# Define the cost function
def compute_cost(X, y, w, b):
    m = len(y)
    predictions = w * X + b
    #-----------↓↓↓-fill your answer-↓↓↓-------------#
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    #-----------↑↑↑-fill your answer-↑↑↑-------------#
    return cost
```

```python
# Define multiple (w, b) combinations for visualization
chosen_params = [
    (5.6, -35),
    (4.0, -40),
    (-1, 20)
]

# Generate values for w and b around chosen values for contour and 3D plot
w_center = 5
b_center = -40
w_range = 30  # Increased range for more complete contours
b_range = 150  # Increased range for better coverage

w_vals = np.linspace(w_center - w_range/2, w_center + w_range/2, 200)  # More points for smoother contours
b_vals = np.linspace(b_center - b_range/2, b_center + b_range/2, 200)  # More points for smoother contours
W, B = np.meshgrid(w_vals, b_vals)

# Compute the cost function for each combination of w and b
Z = np.array([compute_cost(X, y, w, b) for w, b in zip(np.ravel(W), np.ravel(B))])
Z = Z.reshape(W.shape)

# Colors for each chosen (w, b)
colors = itertools.cycle(['blue', 'green', 'orange'])

# Create Matplotlib figure for the first two plots
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Data points with the lines corresponding to chosen (w, b)
ax[0].scatter(X, y, marker='x', color='red', label='Data Points')
x_vals = np.linspace(X.min(), X.max(), 100)

for (w_chosen, b_chosen), color in zip(chosen_params, colors):
    y_vals = w_chosen * x_vals + b_chosen
    ax[0].plot(x_vals, y_vals, color=color, label=f'Line: w={w_chosen}, b={b_chosen}')

ax[0].set_xlabel('Education')
ax[0].set_ylabel('Income')
ax[0].set_title('Data with Chosen Lines')
ax[0].legend()

# Reset color cycle for consistent colors in subsequent plots
colors = itertools.cycle(['blue', 'green', 'orange'])

# Plot 2: Contour plot of the cost function
contour = ax[1].contour(W, B, Z, levels=np.logspace(0, 4, 30), cmap='viridis')  # More levels for better visibility
for (w_chosen, b_chosen), color in zip(chosen_params, colors):
    ax[1].plot(w_chosen, b_chosen, 'o', color=color, markersize=12, label=f'Point: w={w_chosen}, b={b_chosen}')  # Larger points

ax[1].set_xlabel('Slope (w)')
ax[1].set_ylabel('Intercept (b)')
ax[1].set_title('Cost Function Contours')
fig.colorbar(contour, ax=ax[1], label='Cost')
ax[1].legend()

plt.tight_layout()
plt.show()

# Create interactive 3D plot for the third graph
fig_interactive = go.Figure()

fig_interactive.add_trace(go.Surface(z=Z, x=W, y=B, colorscale='Viridis', opacity=0.8))

# Reset color cycle for consistent colors in the 3D plot
colors = itertools.cycle(['blue', 'green', 'orange'])

# Add scatter points for each chosen (w, b)
for (w_chosen, b_chosen), color in zip(chosen_params, colors):
    fig_interactive.add_trace(go.Scatter3d(
        x=[w_chosen],
        y=[b_chosen],
        z=[compute_cost(X, y, w_chosen, b_chosen)],
        mode='markers',
        marker=dict(size=12, color=color),
        name=f'Point: w={w_chosen}, b={b_chosen}'
    ))

# Update layout for better visuals
fig_interactive.update_layout(
    title="Interactive 3D View of the Cost Function",
    scene=dict(
        xaxis_title='Slope (w)',
        yaxis_title='Intercept (b)',
        zaxis_title='Cost J(w, b)'
    ),
    margin=dict(l=0, r=0, b=0, t=50)
)

fig_interactive.show()
```

```python
# Create a matrix of ones and education values
X = np.column_stack([np.ones(len(education)), education])

# Solve the least squares problem
coefficients = np.linalg.lstsq(X, income, rcond=None)[0]

# Print the coefficients
print(coefficients)

# Visualize the results
import matplotlib.pyplot as plt

plt.scatter(education, income, color='blue',label='Actual data')
plt.plot(education, X.dot(coefficients), color='red',label='Linear regression fit')
plt.title('Least Square Fit: Education vs Income')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('income.csv')

# Extract the education and income columns
# X = data['Education']  #this would give error
# Y = data['Income']      #this would give error
X = data['Education'].values.reshape(-1, 1)
Y = data['Income'].values
```

```python
print(X.shape)
print(Y.shape)
```

```python
# Create a linear regression model
model = LinearRegression()

# Fit the model with the data
#-----------↓↓↓-fill your answer-↓↓↓-------------#
model.fit(X, Y)
#-----------↑↑↑-fill your answer-↑↑↑-------------#

# Make predictions
income_pred = model.predict(X)

# Compute performance metrics
mae = mean_absolute_error(Y, income_pred)
mse = mean_squared_error(Y, income_pred)
r2 = r2_score(Y, income_pred)
```

```python
# Plotting the results
plt.scatter(X, Y, color='blue', label='Actual data')
plt.plot(X, income_pred, color='red', linewidth=3, label='Linear regression fit')
plt.xlabel('Education')
plt.ylabel('Income')
plt.title('Linear Regression: Education vs Income')
plt.legend()
plt.show()

# Print the coefficients and performance metrics
print(f"Coefficient (slope): {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")
```
